1) O índice de massa corporal (IMC) é uma medida que usa sua altura e peso para saber se seu peso é saudável. O IMC leva em consideração as variações naturais na forma do corpo, dando uma faixa de peso saudável para uma determinada altura. Calcule o Índice de Massa Corporal. Solicite o nome, idade, peso (kilos) e altura (metros). A formula é dada por: IMC = peso / altura^2 

```python
nome = input("Digite seu nome ")
idade = int(input("Digite sua idade "))
peso  = float(input("Digite o seu peso "))
altura = float(input("Digite a sua altura "))
```

```python
IMC = peso / (altura*altura)
```

```python
print("Seu IMC é : " , IMC)
```

2) Construa um programa em Python que calcule o salário líquido de um professor. Para isso, será necessário que o usuário forneça alguns dados, tais como: valor da hora aula, número de horas trabalhadas e desconto percentual do INSS. Primeiro deve-se estabelecer o valor do salário bruto, efetuar o desconto e obter o salário líquido. Ao final apresente o salário bruto, líquido e desconto.


```python
qnt_hrs = int(input('Informe a quantidade de horas trabalhadas: '))
val_hrs = float(138.41)
des_inss = float(7.5)
```

```python

sal_brt = qnt_hrs * val_hrs
des = (des_inss * sal_brt) / 100
sal_lqd = sal_brt - des
```

```python
print('Seu salario liquido é: R$ %.2f'  % sal_lqd )
print('Seu salário bruto é: R$ % .2f'  % sal_brt)
print('Seu desconto foi de: %.2f'  % des)
```

4) Construa um programa que calcule o volume de um copo da seguinte forma:
Declare as variáveis para raio, altura e volume
Solicite ao usuário o raio e a altura do copo 
Calcule o valor do volume e armazene-o na sua respectiva variável

```python
raio = float(input("Informe o valor do raio: "))
altura = float(input('Informe o valor da altura: '))
```

```python
volume = float(3.14 * (raio ** 2) * altura)
```

```python
print('O valor do volume é: %.2f metros quadrados'   % volume)
```

5) Escreva um programa que pergunte a quantidade de km percorridos por um carro alugado pelo usuário, assim como a quantidade de dias pelos quais o carro foi alugado. Calcule o preço a pagar, sabendo que o carro custa 60,00 reais por dia e 0,15 centavos por km rodado. 

```python
km_per = float(input('informe a distancia percorrida pelo automóvel: '))
qtd_dias = int(input('Informe o tempo do aluguél do automóvel em dias: '))
val_d = 60.00
val_km = 0.15
```

```python
custo_dia = val_d * qtd_dias
custo_km = val_km * km_per
custo_geral = custo_dia + custo_km
```

```python
print('O preço a pagar é: R$%.2f'    %custo_geral)
```
